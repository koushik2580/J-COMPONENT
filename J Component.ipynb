```python
import pandas as pd 
import random as r
import numpy as np
d=pd.read_csv('/content/Salary_Data.csv')
y=d['YearsExperience']
x=d['Salary'] 
n=len(y) 
yp=[]
MSE=[]
def mse(b0,b1):
    for i in range (0,n):
        ypre=b0+(b1*x[i])
        dif=(y[i]-ypre)*(y[i]-ypre)
        yp.append(dif)
        ypp=np.array(yp)
    Mean_SE=np.mean(ypp)
    MSE.append(Mean_SE)
    print("The Mean Square Error for b0=",b0,"b1=",b1,"is: ",Mean_SE)
for i in range(0,150):
  mse(r.randint(0,50),r.randint(0,50))
print("TrE Least Mean Square Error is",min(MSE))
```
